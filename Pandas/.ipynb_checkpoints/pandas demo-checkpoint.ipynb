```python
p_df[filt]
```

```python
p_df.loc[filt] # using filter with loc
```

```python
p_df.loc[filt, 'last'] # filtering with .loc, filter and column
```

```python
filt1 = (p_df['last'] == 'john') & (p_df['first'] == 'faith')  # AND operator
```

```python
filt2 = (p_df['last'] == 'john') | (p_df['first'] == 'faith')  # OR operator
```

```python
p_df.loc[-filt] # get things that do not match a filter use - before filter
```
