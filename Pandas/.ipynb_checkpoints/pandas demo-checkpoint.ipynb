```python
people_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_respondents = df['Country'].value_counts()
python_df = pd.concat([country_respondents, people_uses_python], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'count':'NumRespondents', 'LanguageWorkedWith':'NumKnowsPython'}, inplace = True)
python_df
```

```python
python_df['PctKnowsPython'] = python_df['NumKnowsPython']/python_df['NumRespondents']*100
python_df
```

```python
python_df.sort_values(by='PctKnowsPython', ascending = False)
```
